```python
from scipy.stats import beta
import seaborn as sns
```

```python
data_beta = beta.rvs(50, 1, size=100000)
sns.distplot(data_beta, kde=False, bins=100)
```

```python
data_beta = beta.rvs(288, 30, size=1000)
sns.distplot(data_beta, kde=False, bins=100)
data_beta.var()
```

```python
data_beta.mean()
```

## Scaled

```python
data_beta = beta.rvs(28, 3, size=1000)
sns.distplot(data_beta, kde=False, bins=100)
data_beta.var()
#data_beta.mean()
```
